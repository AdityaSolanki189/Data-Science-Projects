```python
canvas = np.zeros((300, 300, 3), dtype = "uint8")
#RGB(255,255,255)
green = (0, 255, 0)
cv.line(canvas, (0, 0), (300, 300), green,4)
cv_imshow(canvas)
cv.waitKey(0)
```

```python
red = (0, 0, 255)
cv.line(canvas, (300, 0), (0, 300), red, 3)
cv_imshow(canvas)
cv.waitKey(0)
```

```python
M = np.float32([[1, 0, 25], [0, 1, 50]])
print(image.shape[0])
print(image.shape[1])
shifted = cv.warpAffine(image, M, (image.shape[1], image.shape[0]))
cv_imshow(shifted)
cv.waitKey(0)
```

```python
# Rotation
cv_imshow(image)
(h, w) = image.shape[:2]

center = (w // 2, h // 2)
M = cv.getRotationMatrix2D(center, 45, 1.0)  #(point of rotation, angle, scale)
rotated = cv.warpAffine(image, M, (w, h))
cv_imshow(rotated)
cv.waitKey(0)
```

```python

cv_imshow(image)
print(image.shape)
r = 150.0 / image.shape[1]
dim = (150, int(image.shape[0] * r))
resized = cv.resize(image, dim, interpolation = cv.INTER_AREA)
print(resized.shape)
cv_imshow(resized)
cv.waitKey(0)
```

```python

cv_imshow(image)
flipped = cv.flip(image, 0)  #0 for flipping horizonatlly
cv_imshow(flipped)
cv.waitKey(0)
```
