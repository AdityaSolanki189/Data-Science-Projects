```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report ,f1_score, plot_roc_curve
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

from sklearn.metrics import plot_confusion_matrix
```

```python
df = pd.read_csv('D:\\Software\\Anaconda\\ANA_Codes\\Data\\breast-cancer.csv')
```

```python
df.describe()
```

```python
df.head(5)
```

```python
y = df.diagnosis
```

```python
#label encoding
lb = LabelEncoder()
df.diagnosis = lb.fit_transform(y)
```

```python
df.head(5)
```

```python
x = df.iloc[:,2:32]
print('X shape',x.shape)
```

```python
scaler = StandardScaler()
x_std = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_std, y, test_size = 0.37, random_state = 0)
```

```python
svc = svm.SVC(C = 0.4, kernel = 'linear')
```

```python
svc.fit(x_train,y_train)
```

```python
print(svc.score(x_train,y_train)*100)
```

```python
print(svc.score(x_test, y_test)*100)
```

```python
plot_confusion_matrix (svc, x_test, y_test)
```

```python
y_pred = svc.predict(x_test)
cr = classification_report(y_test, y_pred)
print(cr)
```

```python
plot_roc_curve(svc, x_test, y_test)
```
