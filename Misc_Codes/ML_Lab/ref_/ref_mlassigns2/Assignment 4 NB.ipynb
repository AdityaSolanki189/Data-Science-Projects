```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/diabetes. csv')    #using datset of diabetes available on kaggle/ ucl
```

```python
df.head()
```

```python
df.info()   #information about the dataset
```

```python
df.isna().sum()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,8]    #the target feature here is the outcome
```

```python
scaler = StandardScaler()       #standardizing the dataset
x_std = scaler.fit_transform(x)
```

```python
#splitting the dataset in a ratio of 70:30

x_train, x_test, y_train, y_test = train_test_split(x_std, y , test_size=0.3, random_state=None)
print('x training shape', x_train.shape)
print('y training shape', y_train.shape)
print('x testing shape', x_test.shape)
print('y testing shape', y_test.shape)
```

```python
#importing naive bayes classifier from sklearn
from sklearn.naive_bayes import GaussianNB 

nb = GaussianNB()   #object of gaussian NB
```

```python
#training the model
nb.fit(x_train, y_train)
```

```python
#accuracy score of training sets

nb.score(x_train, y_train)
```

```python
#accuracy score of testing sets

nb.score(x_test, y_test)
```
