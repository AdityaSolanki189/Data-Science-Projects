```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
lbl = LabelEncoder()
lbl.fit_transform(df.ocean_proximity)
```

```python
df.isna().sum()
```

```python
emi = SimpleImputer(strategy='mean', missing_values=np.nan)
df['total_bedrooms'] = emi.fit_transform(df[['total_bedrooms']])
```

```python
#df = df.dropna(axis='index', how='any')
#index goes row wise
#column will delete the whole column
#any or all

#fillna(inplace=True)

#drop the entire column
#df.drop('ocean_proximity', axis='column', inplace=True)
```

```python
df.drop('ocean_proximity', axis='columns', inplace=True)
df
```

```python
x = df.iloc[:,0:8]
y = df.iloc[:,8]

scaler = StandardScaler()
x_std = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_std, y , test_size=0.3, random_state=None)
print('x training shape', x_train.shape)
print('y training shape', y_train.shape)
print('x testing shape', x_test.shape)
print('y testing shape', y_test.shape)
```

```python
mlp = MLPRegressor(hidden_layer_sizes=(100,50), activation='relu', solver="lbfgs", max_iter=20)
mlp
```

```python
history = mlp.fit(x_train, y_train)
```

```python
#accuracy score for training set
print(mlp.score(x_train,y_train)*100)
```

```python
#accuracy score for testing set
print(mlp.score(x_test,y_test)*100)
```

```python
y_pred = mlp.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score, mean_squared_log_error

print(r2_score(y_test, y_pred))
print(mean_squared_log_error(y_test, y_pred))
```

```python
df_temp = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df_temp.head()
```

```python
df_temp = df_temp.head(30)
df_temp.plot(kind='bar',figsize=(10,6))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python

plt.plot(mlp.loss_curve_)
plt.title("Loss Curve", fontsize=14)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```
