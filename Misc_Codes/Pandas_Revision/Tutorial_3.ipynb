Different ways of creating a DataFrame.

```python
import pandas as pd

# Importing CSV
df = pd.read_csv('.//datasets//weather_data.csv')
df
```

```python
# Importing Excel
df2 = pd.read_excel('.//datasets//SampleData.xlsx')
df2.head(5)
```

```python
# Creating DF using a Python Dictionary
weather_data = {
    'day': ['1/1/2017', '1/2/2017', '1/3/2017'],
    'temperature': [32, 35, 28],
    'windspeed': [6, 7, 2],
    'event': ['Rain', 'Sunny', 'Snow']
}

df3 = pd.DataFrame(weather_data)
df3
```

```python
# Creating DF using Tuples List
weather_data2 = [
    ('1/1/2017', 32, 6, 'Rain'),
    ('1/2/2017', 35, 7, 'Sunny'),
    ('1/3/2017', 28, 2, 'Snow')
]

df4 = pd.DataFrame(weather_data2, columns=['day', 'temperature', 'windspeed', 'event'])
df4
```

```python
# Creating DF using a List of Dictionaries
weather_data3 = [
    {'day': '1/1/2017', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2017', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2017', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'},
]

df5 = pd.DataFrame(data=weather_data, columns=['day','temperature','windspeed','event'])
df5
```
